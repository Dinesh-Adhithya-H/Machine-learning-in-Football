```
#find area of triangle
def area_triangle(x1, y1, x2, y2, x3, y3): 
  
    return abs((x1 * (y2 - y3) + x2 * (y3 - y1)  
                + x3 * (y1 - y2)) / 2.0) 
# A function to check whether point P(x, y) 
# lies inside the triangle formed by  
# A(x1, y1), B(120, 36) and C(120, 44)  
# B and C are the coordinates for the goalposts 
def isInside(x1, y1, x, y): 
  
    # Calculate area of triangle ABC 
    A = area_triangle(x1, y1, 120, 36, 120, 44) 
  
    # Calculate area of triangle PBC  
    A1 = area_triangle(x, y, 120, 36, 120, 44) 
      
    # Calculate area of triangle PAC  
    A2 = area_triangle(x1, y1, x, y, 120, 44) 
      
    # Calculate area of triangle PAB  
    A3 = area_triangle(x1, y1, 120, 36, x, y) 
      
    # Check if sum of A1, A2 and A3  
    # is same as A 
    if(A == A1 + A2 + A3): 
        return True
    else: 
        return False
```

```
opponent_in_triangle=[]
teammate_in_triangle=[]
g1=[120,36]
g2=[120,44]
for i in range(len(data)):
  o=0
  t=0
  for j in data["players_home"].iloc[i]:
    if isInside(data["location"].iloc[i][0],data["location"].iloc[i][1],j[0],j[1]):
      t+=1
  for j in data["players_away"].iloc[i]:
    if isInside(data["location"].iloc[i][0],data["location"].iloc[i][1],j[0],j[1]):
      o+=1
  opponent_in_triangle.append(o)
  teammate_in_triangle.append(t)
```
